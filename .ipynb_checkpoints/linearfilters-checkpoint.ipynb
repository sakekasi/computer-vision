# Linear Filters

```python
%matplotlib inline 

from scipy import ndimage
from scipy import misc
import numpy as np
import matplotlib.pyplot as plt
```

messing around with images

```python
f = ndimage.imread("cat.jpg")
plt.imshow(f)

print f.shape
```

trying out some prebuilt functions:

* a gaussian

```python
plt.imshow(ndimage.gaussian_filter(f, 10))
```

* a box filter

```python
k = 50

plt.imshow(ndimage.filters.generic_filter(f, lambda arr: (arr.sum())/(k**2),
                                          size=(k, k, 1), mode='constant'))
```

* a sobel filter

```python
k = 3

def sobel(arr):
    filter = [-1, 0, 1,
              -2, 0, 2,
              -1, 0, 1]
    return np.multiply(arr, filter).sum()/8
    

plt.imshow(ndimage.filters.generic_filter(f, sobel,
                                          size=(k, k, 1), mode='constant'))
```
